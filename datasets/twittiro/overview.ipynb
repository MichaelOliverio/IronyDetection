```python
import conllu
import pandas as pd
```

```python
with open("it_twittiro-ud-train.conllu", "r", encoding="utf-8") as f:
    data = conllu.parse(f.read())
```

```python
# Loop tra le frasi
for i, sentence in enumerate(data):
    print(f"\n=== Frase {i+1} ===")
    
    # Stampa i metadati (es. sent_id, text, twittiro)
    if sentence.metadata:
        for key, value in sentence.metadata.items():
            print(f"{key}: {value}")
    
    # Convertila in DataFrame per leggibilità
    df = pd.DataFrame(sentence)

    break

df
```
